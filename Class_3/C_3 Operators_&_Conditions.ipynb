# Arthmetic Operators

1.   Addition ( + )
2.   Subtraction ( - )
3.   Multiplication ( * )
4.   Division ( / )
5.   Remainder ( % )



```python
faculty_mem = 12
staff = 4
students = 100
absent = 15
total = faculty_mem + staff + students - absent
rabri = 250

rabri_result = rabri * total

result = f"chand tara need to make {rabri_result}g of rabri"
print(result)

kg_result = f"chand tara need to make {rabri_result // 1000}kg of rabri"
print(kg_result)
```

# Conditions

*   if & else

## Conditinal Operators

1.   Equals ( == )
2.   Not Equals ( != )
3.   Greater Than ( > )
4.   Less Than ( < )
3.   Greater Than or Equals ( >= )
4.   Less Than or Equals ( <= )

## Logical Operators

1.   And ( and )
2.   Or ( or )
3.   Not ( not )


```python
# Simple Conditions
age = 18
my_age = 18

if(my_age >= age):
  print("you are allowed")
else:
  print("you are not allowed")
```

```python
# Conditions with Operators

age = int(input("how old are you: "))

if( age < 18):
  print("you are a minor")
elif( age >= 18 and age < 40 ):
  print("you are a young person")
elif( age >= 40 and age < 50   ):
  print("you are a uncle")
elif( age >= 50):
  print("you are an old person")
else:
  print("you are a human")
```
